```python
df_cleaned = df.dropna(subset=['Medicine Name', 'Brand', 'Price'])

def get_medicine_price(medicine_name, brand_name):

    result = df_cleaned[
        (df_cleaned['Medicine Name'].str.contains(medicine_name, case=False, na=False)) &
        (df_cleaned['Brand'].str.contains(brand_name, case=False, na=False))
    ]

    if not result.empty:
        price = result['Price'].iloc[0]
        return f"The price of '{medicine_name}' by '{brand_name}' is {price}."
    else:
        return "Medicine or brand not found in the dataset."

medicine_name_input = input("Enter the medicine name: ").strip()
brand_name_input = input("Enter the brand name: ").strip()

output = get_medicine_price(medicine_name_input, brand_name_input)
print(output)
```
